# Notebook 1 — Data Cleaning
### Crash Risk Intelligence System
**Dataset:** Maryland Crash Reporting — Drivers Data  
**Records:** 215,034 rows × 39 columns  
**Goal:** Clean raw crash data and prepare it for EDA and feature engineering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)

print("Libraries imported successfully")
```

```python
# Load full dataset
df = pd.read_csv('C:\crash_risk_intelligence\data\Crash_Reporting.csv', low_memory=False)

print(f"Rows: {df.shape[0]:,}")
print(f"Columns: {df.shape[1]}")
```

```python
# See first 5 rows
df.head()
```

```python
# Calculate missing value percentage for each column
missing = (df.isnull().sum() / len(df) * 100).round(1)
missing = missing[missing > 0].sort_values(ascending=False)

print("Columns with missing values:\n")
print(missing)
```

## Step 1 — Drop High Null % Columns
Columns with more than 40% missing values are not useful for analysis or modeling.
Dropping 5 columns:
- Non-Motorist Substance Abuse (97.3%)
- Related Non-Motorist (96.7%)
- Municipality (91.1%)
- Off-Road Description (90.7%)
- Circumstance (80.5%)

```python
# Drop columns with more than 40% missing values
cols_to_drop = [
    'Non-Motorist Substance Abuse',
    'Related Non-Motorist',
    'Municipality',
    'Off-Road Description',
    'Circumstance'
]

df.drop(columns=cols_to_drop, inplace=True)

print(f"Columns dropped: {len(cols_to_drop)}")
print(f"New shape: {df.shape}")
```

## Step 2 — Fix Inconsistent Casing
Several columns have the same value written in different cases.
For example 'Clear' and 'CLEAR', 'Unknown' and 'UNKNOWN'.
Standardizing all text columns to Title Case.

```python
# Standardize all text/object columns to Title Case
text_cols = df.select_dtypes(include='object').columns

for col in text_cols:
    df[col] = df[col].str.strip().str.title()

# Verify fix on Weather column
print("Weather unique values after fix:")
print(df['Weather'].value_counts())
```

## Step 3 — Fix Crash Date/Time Column
'Crash Date/Time' is currently stored as a string/text.
Converting it to proper datetime format so we can 
extract hour, day, month etc. in Feature Engineering.

```python
# Convert Crash Date/Time to datetime
df['Crash Date/Time'] = pd.to_datetime(df['Crash Date/Time'], errors='coerce')

print("Data type after conversion:", df['Crash Date/Time'].dtype)
print("\nSample values:")
print(df['Crash Date/Time'].head())
```

## Step 4 — Handle Remaining Missing Values
For columns with moderate missing values we use context-aware filling:
- Categorical columns → fill with 'Unknown' (preserves real distribution)
- Numerical columns → fill with median (robust to outliers)
We do NOT use mean for categorical columns as it distorts distributions.

```python
# Fill categorical columns with 'Unknown'
cat_cols = df.select_dtypes(include='object').columns
df[cat_cols] = df[cat_cols].fillna('Unknown')

# Fill numerical columns with median
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

# Verify no missing values remain
remaining_nulls = df.isnull().sum().sum()
print(f"Total missing values remaining: {remaining_nulls}")
```

## Step 5 — Remove Duplicates
Checking for and removing exact duplicate rows.
Duplicate rows can skew analysis and model training
by giving certain crashes more weight than others.

```python
# Check duplicates before
print(f"Duplicate rows before: {df.duplicated().sum()}")

# Drop duplicates
df.drop_duplicates(inplace=True)

# Check duplicates after
print(f"Duplicate rows after: {df.duplicated().sum()}")
print(f"Final shape: {df.shape}")
```

## Step 6 — Fix Mixed Type Columns
'Local Case Number' had mixed types (numbers and strings).
Converting it to string since it is an ID column, not used for math.

```python
# Convert ID-like columns to string
df['Local Case Number'] = df['Local Case Number'].astype(str)

print("Local Case Number dtype:", df['Local Case Number'].dtype)
print("Sample values:")
print(df['Local Case Number'].head())
```

## Step 7 — Save Cleaned Data
Saving the cleaned dataframe as a new CSV file for use in the next notebook.
We also save a 1000 row sample for GitHub since the full file is too large to push.

```python
# Save full cleaned data for local use
df.to_csv('C:\crash_risk_intelligence\data\cleaned_crash_data.csv', index=False)
print(f"Full cleaned data saved — shape: {df.shape}")

# Save small sample for GitHub
df.sample(1000, random_state=42).to_csv('C:\crash_risk_intelligence\data\sample_data.csv', index=False)
print("Sample data saved — 1000 rows")
```
